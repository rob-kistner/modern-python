## Dicts: Iterating

```python
instructor = {
    "name": "Colt",
    "owns_dog": True,
    "num_courses": 4,
    "favorite_language": "Python",
    "is_hilarious": False,
    44: "my favorite number"
}

instructor
```

### keys()

Loop over keys


```python
for k in instructor.keys():
    print(f"{k}")
```

### values()

Loop over values


```python
for v in instructor.values():
    print(f"{v}")
```

### items()

Loop over key/value as tuples, should use two vars (k,v) to separate the keys and values in the loop.


```python
for k, v in instructor.items():
    print(f"{k}: {v}")
```

## Exercise 29: Dictionary Access


```python
from random import choice

food = choice(["cheese pizza", "quiche","morning bun","gummy bear","tea cake"])

bakery_stock = {
    "almond croissant" : 12,
    "toffee cookie": 3,
    "morning bun": 1,
    "chocolate chunk cookie": 9,
    "tea cake": 25
}

def in_stock(food):
    if food in bakery_stock:
        print(f"{bakery_stock.get(food)} left")
    else:
        print("We don't make that")
```

```python
in_stock('almond croissant')
```

```python
in_stock('morning bun')
```

```python
in_stock('gummies')
```

# Dicts: Comprehension

```python
numbers = dict(first=1, second=2, third=3)

instructor = {
    "name": "Colt",
    "favorite_language": "Python",
    "44": "my favorite number",
    "color": "purple"
}

print(numbers)
print(instructor)
```

Square the numbers

```python
{ key: value ** 2 for key, value in numbers.items() }
```

Key is the number, value is the square

```python
{num: num**2 for num in [1,2,3,4,5]}
```

Created from two strings

```python
str1 = "ABC"
str2 = "123"

{str1[i]: str2[i] for i in range(0, len(str1))}
```

Capitalize all keys & values

```python
{k.upper():v.upper() for k, v in instructor.items()}
```

Interpret even or odd number into a dict

```python
num_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]
# num_list = range(1, 100)

{ num:("even" if num%2 == 0 else "odd") for num in num_list }
```

Only cap a key if it matches a certain key

```python
{
    (k.upper() if k == "color" else k):v.upper() 
    for k, v in instructor.items()
}
```
