## Tuples: Looping

```python
months = ('January','February','March','April','May','June','July','August','September','October','November','December')

nums = (1, 2, 3, 3, 3)
```

Use exact same syntax as a list

```python
for month in months:
    print(month)
```

### count()

Show number of occurances in a tuple

```python
nums.count(3)
```

### index()

Same as a list

```python
nums.index(3)
```

Can have nested tuples

```python
nested_nums = (1, 2, 3, (4, 5), 6, 7)

nested_nums[3]
```

```python
nested_nums[3][0]
```
